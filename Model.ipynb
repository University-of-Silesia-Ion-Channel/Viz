```python
import numpy as np
import matplotlib.pyplot as plt

class IonModel:
    def __init__(self):
        self.x = np.array([], dtype=np.float64)
        self.t = np.array([])

    def modelEuler(self, a, b, x0, delta_t=0.001, records=1000):
        # definicja funkcji siły
        self.F = lambda x, a, b: -((a * x*x*x) - (b * x))
        # self.F = lambda x, a, b: -x
        # self.F = lambda x, a, b: -a*x

        # dodanie pierwszego elementu do tablicy
        if np.size(self.x) == 0:
            self.x = np.append(self.x, x0)

        # wypełnienie czasów
        self.t = np.arange(0, records * delta_t, delta_t)

        t = 1
        while t < records:
            # print(self.x[t - 1] + self.F(self.x[t - 1], a, b) * delta_t)
            self.x = np.append(
                self.x, self.x[t - 1] + self.F(self.x[t - 1], a, b) * delta_t
            )
            t += 1
        plt.plot(self.t, self.x)
        plt.show()
```

```python
IM = IonModel()
IM.modelEuler(1, 1, 7, 0.001, 1000)
```

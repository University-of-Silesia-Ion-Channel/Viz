```python
import pandas as pd
import plotly.express as px
import os
class IonVisualisation:

    def __init__(self):
        pass

    def visualize(self, batch_size):
        assert isinstance(batch_size, int), "batch_size must be an integer"
        assert batch_size > 0, "batch_size must be greater than 0"
        # read all the files in a specified directory
        data_dir = os.path.join(os.getcwd(), "dane", "20mV")
        data_file_names = self._list_files(data_dir)
        files_data = []
        for file in data_file_names:
            cnt = 0
            sum = 0
            file_data = []
            with open(os.path.join(data_dir, file)) as data_file:
                for record in data_file:
                    if cnt < batch_size:
                        record = record.strip()
                        sum += float(record)
                        cnt+=1
                    else:
                        cnt = 0
                        file_data.append(sum / batch_size)
                        sum = 0
            files_data.append(file_data)
            fig = px.scatter(file_data)
            fig.update_layout(
            title="Plot Title",
            xaxis_title="time",
            yaxis_title="voltage",
            font=dict(
            family="Courier New, monospace",
            size=12,
            color="RebeccaPurple"))
            fig.show()
        # px.scatter(files_data[3])
        
    
    def _list_files(self, path):
        onlyfiles = [f for f in os.listdir(path) if os.path.isfile(os.path.join(path, f))]
        return onlyfiles
```

```python
iv = IonVisualisation()
iv.visualize(1000)
```
